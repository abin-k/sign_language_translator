**train test split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.3, random_state=12)
```

```python
img_size=x[0].shape
img_size
```

```python
x_train = np.array(x_train) / 255
x_test = np.array(x_test) / 255

x_train = x_train.reshape(-1, img_size[0], img_size[1], img_size[2])
y_train = np.array(y_train)

x_test = x_test.reshape(-1, img_size[0], img_size[1], img_size[2])
y_test = np.array(y_test)
```

```python

model = Sequential()
#convolutional layer 1
model.add(Conv2D(filters=128,kernel_size=(11,11),strides=(4,4),activation='relu',input_shape=(img_size[0], img_size[1], img_size[2])))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

#convolutional layer 2
model.add(Conv2D(filters=256,kernel_size=(5,5),strides=(1,1),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(3,3)))


#convolutional layer 3
model.add(Conv2D(filters=256,kernel_size=(3,3),strides=(1,1),activation='relu',padding='same'))
model.add(BatchNormalization())


#convolutional layer 4
model.add(Conv2D(filters=256,kernel_size=(1,1),strides=(1,1),activation='relu',padding='same'))
model.add(BatchNormalization())


#convolutional layer 5
model.add(Conv2D(filters=256,kernel_size=(1,1),strides=(1,1),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

#flattern layer
model.add(Flatten())

#FCL 1
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.5))

#FCL 1
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.5))

#FCL 2
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.5))

#FCL 3
model.add(Dense(10,activation='softmax'))
```

**Real time prediction**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from tensorflow import keras


# Load the saved model
#model = keras.models.load_model('/content/drive/MyDrive/sign_language_model.h5')

# Define a dictionary to map output classes to letters
classes = {0: '0', 1: '1', 2: '2', 3: '3',4:'4',5:'6',7: '7', 8: '8', 9: '9'}

# Start the video stream
cap = cv2.VideoCapture(0)

while True:
    # Capture a frame
    ret, frame = cap.read()
    
    if not ret:
        continue
    
    # Preprocess the frame
    frame = cv2.resize(frame, (224, 224))
    frame = np.expand_dims(frame, axis=0)
    frame = frame / 255.0
    
    # Make a prediction
    pred = model.predict(frame.reshape(-1,400,400,3))
    letter = classes[np.argmax(pred)]
    
    # Display the letter on the screen
    cv2.putText(frame, letter, (50,50), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,0), 2)
    cv2_imshow(frame)
    
    # Exit the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video stream and close the window
cap.release()
cv2.destroyAllWindows()
```
